#Punto 1

```
valorPeso = None
while (valorPeso == None):
  try:
    valorPeso = float(input('Ingrese un valor en COP que desee pasar a dolares: '))
  except ValueError:
    print('El valor ingresado no es valido')
valorDolar = round(valorPeso/3800,2)
print (f'El valor en dolares es {valorDolar}')
```

#Punto 3

```
import matplotlib.pyplot as plt

meses = ['Enero', 'Febrero', 'Marzo', 'Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
ingresos = [7000000,8000000,8500000,75000000,6700000,7200000,8400000,8500000,75000000,6900000,8350000,8000000]
plt.bar (meses,ingresos)
plt.title('Ingresos vs Meses')
plt.xlabel('Meses')
plt.ylabel('Ingresos')
plt.show()
```

```
import matplotlib.pyplot as plt
ciudades = ['Bogotá','Medellín','Cali','Barranquilla','Cartagena']
areas = [1775,385,619,154,609.1]
explodeList = [0.2,0,0,0,0]

plt.title('Area de 5 ciudades en Colombia')
plt.pie(areas,labels = ciudades, explode=explodeList, shadow = True,startangle=60)
plt.show()
```

```
import pandas as pd
ecg = pd.read_csv('ecg (1).csv',encoding='UTF-8', header=0, delimiter=';').to_dict()
ecg.keys()
x = list (ecg['muestra'].values())
y = list (ecg['valor'].values())
import matplotlib.pyplot as plt
plt.plot(x,y,'r')
plt.title('Grafica ECG')
plt.xlabel('Tiempo(ms)')
plt.ylabel ('Voltaje (uV)')
plt.savefig('ecg.png')
plt.show()
```
